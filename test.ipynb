```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from random import shuffle
import glob
from sklearn.preprocessing import LabelEncoder as le
from sklearn.model_selection import train_test_split
import time
import keras
from keras.models import Sequential,load_model
from keras.layers import Dense, Dropout, Flatten, GlobalAveragePooling2D
from keras.layers import Conv2D, MaxPooling2D
from keras.applications import MobileNet
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Model
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping, ModelCheckpoint
import pickle
from keras.utils import plot_model


data = []


kk = glob.glob('TestData/*.jpg')

pickle_in = open("dict_c.pickle","rb")
dict_c = pickle.load(pickle_in)


pickle_in = open("dict_f_20_early_stopping_90.pickle","rb")
dict_f = pickle.load(pickle_in)
```

```python
for i in range(len(dict_f)):
    k,lk = dict_f[i].split('_')
    dict_f[i] = k + '@'+ lk
```

```python
model_c = load_model('CO_15.h5')
model_f = load_model('FINE_20_early_stopping_90.h5')
```

```python
kk[0]
```

```python
t = time.time()
y_test = []
for i in kk:
    po = mpimg.imread(i)/255.0
    po = np.resize( cv2.resize(po,(224,224)),(1,224,224,3))
    pc = model_c.predict(po)
    pf = model_f.predict(po)
    pca = np.argmax(pc)
    pfa = np.argmax(pf)
    pl = i.split('/')
    y_test.append([pl[1], dict_c[pca], dict_f[pfa]])
    print(dict_c[pca], dict_f[pfa])
print(time.time() - t)
```
